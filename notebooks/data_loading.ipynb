# Schema Differences
| Dataset | Key Fields                       | Data Type       | Notes                            |
|---------|----------------------------------|----------------|----------------------------------|
| NYC 311 | Unique Key                        | object         | primary identifier               |
| NYC 311 | Created Date / Closed Date        | object         | dates as strings, may need parsing |
| NYC 311 | Problem (formerly Complaint Type) | object         | categorical, inconsistent casing |
| NYC 311 | Problem Detail (formerly Descriptor) | object      | free-form text                   |
| NYC 311 | Latitude / Longitude             | float64        | numeric, missing values possible |
| Yelp    | business_id                       | object         | primary identifier               |
| Yelp    | categories                        | object         | comma-separated categories       |
| Yelp    | review text (not listed in dtypes)| object         | long text, unstructured          |
| Yelp    | latitude / longitude              | float64        | numeric, geographic coordinates  |
| Yelp    | stars                             | float64        | numeric, rating                  |
