```python
import pandas as pd
import json

#NYC 311
nyc_311 = pd.read_csv("../data/raw/nyc_311/nyc_311_raw.csv", low_memory=False)

yelp_business = pd.read_json(
    "../data/raw/yelp/yelp_academic_dataset_business.json",
    lines = True
)
```

```python
def standardize_columns(df):
    df.columns = (
        df.columns
        .str.lower()
        .str.strip()
        .str.replace(" ", "_")
        .str.replace(r"[^\w_]", "", regex = True)
    )
    return df
nyc_311 = standardize_columns(nyc_311)
yelp_business = standardize_columns(yelp_business)

nyc_311.to_csv("../data/processed/nyc_311_cleaned.csv", index = False)
yelp_business.to_csv("../data/processed/yelp_business_cleaned.csv", index = False)
```
