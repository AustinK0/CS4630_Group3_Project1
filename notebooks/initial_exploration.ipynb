```python
import pandas as pd
import json

#Philly 311
philly_311 = pd.read_csv("../data/raw/philly_311_raw.csv", low_memory=False)

yelp_business = pd.read_json(
    "../data/raw/yelp_academic_dataset_business.json",
    lines = True
)
```

```python
def standardize_columns(df):
    df.columns = (
        df.columns
        .str.lower()
        .str.strip()
        .str.replace(" ", "_")
        .str.replace(r"[^\w_]", "", regex = True)
    )
    return df
philly_311 = standardize_columns(philly_311)
yelp_business = standardize_columns(yelp_business)

philly_311.to_csv("../data/processed/philly_311_cleaned.csv", index = False)
yelp_business.to_csv("../data/processed/yelp_business_cleaned.csv", index = False)
```
